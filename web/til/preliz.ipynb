````python
import arviz as az
import preliz as pz
az.style.use("arviz-white")
````

## Exploring parameters in a Beta distribution

1. **$\alpha$ (Alpha)**: This parameter represents the number of successes plus one in the context of binomial trials. It can be thought of as the prior belief about the number of successes observed.

2. **$\beta$ (Beta)**: This parameter represents the number of failures plus one. It reflects the prior belief about the number of failures observed.

Together, $\alpha$ and $\beta$ shape the distribution and reflect the prior knowledge or belief about the probability of success in a binomial experiment.

````python
# Example plot of the beta probabilty density function equivalent to assuming you will 2 successes out of 7 trials.
pz.Beta(2+1,5+1).plot_pdf(pointinterval=True)
````
