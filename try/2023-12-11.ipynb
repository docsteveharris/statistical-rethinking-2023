```python
import numpy as np
import polars as pl
from scipy import stats as stats
import plotly as plt
```

```python
a,b = 6,3
beta_posterior = stats.beta(a,b)
samples = beta_posterior.rvs(size=10000)
```

```python
df = pl.DataFrame(samples)
```

```python
print(df)
```

```python
df['column_0']
```

```python
import plotly.graph_objects as go
fig = go.Figure(data=go.Histogram(x=df["column_0"]))
fig.show()
```
